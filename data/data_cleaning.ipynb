### Store Data
* Double check all the unique value in each column in store data
    * Issue: Unique Values in PromoInterval are **[nan 'Jan,Apr,Jul,Oct' 'Feb,May,Aug,Nov' 'Mar,Jun,Sept,Dec']**, which is not easy to read, so it transfer to:
        * nan to 0
        * 'Jan,Apr,Jul,Oct' to 1
        * 'Feb,May,Aug,Nov' to 2 
        * 'Mar,Jun,Sept,Dec' to 3

* Issue: Too many null values in the store data
    * ❌ Solve 1: If it use **dropna** function, it will only keep 365 rows(stores) which means only **32.74% rows remain after dropna**.
    * Solve 2: transfer the values separately
        * use four columns (Promo2 / Promo2SinceWeek / Promo2SinceYear / PromoInterval) to feature new column **Promo2Rounds** which means how many **times** have the stores run the promotion until the end of year 2015.
            * use the end of 2015 as the anchor due to some stores in the dataset were temporarily closed for refurbishment.

```python
cleaned_store_data = store_data.copy()

# what will happen if dropna in store_data
if_dropna_store_data = cleaned_store_data.dropna()
exist_rows = if_dropna_store_data.shape[0]
exist_percentage = exist_rows / cleaned_store_data.shape[0] * 100
print(f"{exist_percentage:.2f}% rows remain after cleaned_store_data.dropna()")
```

```python

# set the end of 2015 as anchor date 
anchor_date = pd.Timestamp(date(2015, 12, 31))

# month mapping to number
MONTH_MAP = {
    "Jan":1, "Feb":2, "Mar":3, "Apr":4, "May":5, "Jun":6,
    "Jul":7, "Aug":8, "Sep":9, "Oct":10, "Nov":11, "Dec":12
}

def iso_week_monday(year, week):
    """
    transfer (ISO year, ISO week) to the Monday date of that week; return NaT for missing or invalid values
    """
    try:
        if pd.isna(year) or pd.isna(week) or int(week) <= 0:
            return pd.NaT
        return pd.Timestamp.fromisocalendar(int(year), int(week), 1)
    except Exception:
        return pd.NaT

def parse_interval(interval):
    """
    transfer 'Feb,May,Aug,Nov' to {2,5,8,11} as example; return empty set for missing or empty string
    """
    if pd.isna(interval):
        return set()
    if not isinstance(interval, str):
        interval = str(interval)
    interval = interval.strip()
    if not interval:
        return set()
    return {MONTH_MAP.get(m.strip()[:3], None) for m in interval.split(",")} - {None}

def count_rounds_to_date(start_ts, months_set, anchor=anchor_date):
    """
    count the number of times from start_ts to anchor that the month is in months_set
    """
    if pd.isna(start_ts) or not months_set:
        return 0
    if anchor < start_ts:
        return 0

    start_ms = pd.Timestamp(start_ts.year, start_ts.month, 1)
    end_ms   = pd.Timestamp(anchor.year, anchor.month, 1)

    month_starts = pd.date_range(start=start_ms, end=end_ms, freq="MS")
    return sum((dt.month in months_set) and (dt >= start_ts) for dt in month_starts)


# transfer (Promo2SinceYear, Promo2SinceWeek) to Promo2StartDate as specific timestamp
cleaned_store_data["Promo2StartDate"] = [
    iso_week_monday(y, w) for y, w in zip(cleaned_store_data["Promo2SinceYear"], cleaned_store_data["Promo2SinceWeek"])
]

# extract the set of months each store starts promotion
months_sets = cleaned_store_data["PromoInterval"].apply(parse_interval)

# compute how many times each store has run Promo2 until anchor_date; set to 0 if none
cleaned_store_data["Promo2Rounds"] = 0  
mask_calc = (cleaned_store_data["Promo2"] == 1)

cleaned_store_data.loc[mask_calc, "Promo2Rounds"] = [
    count_rounds_to_date(start_ts, mset, anchor=anchor_date)
    for start_ts, mset in zip(cleaned_store_data.loc[mask_calc, "Promo2StartDate"], months_sets[mask_calc])
]

cleaned_store_data.drop(columns=["Promo2SinceYear", "Promo2SinceWeek", "Promo2StartDate", "PromoInterval"], inplace=True)
```
